```python
rng = np.random.default_rng(5)
# uniformly distributed in the range [0, 1), then shifted to the range [-0.5, 0.5) by subtracting 0.5.
x1 = rng.uniform(size=500) - 0.5
x2 = rng.uniform(size=500) - 0.5
# Boolean mask Labels each point based on whether x1**2 > x2**2, i.e., whether it lies in the region where the square of the x-coordinate 
#is greater than the square of the y-coordinate.
y = x1**2 - x2**2 > 0
```

```python
fig , ax = plt.subplots(figsize=(8,8))
ax.scatter(x1,
x2,
c=y,
cmap=plt.cm.coolwarm)
ax.set_xlabel('Feature x1')
ax.set_ylabel('Feature x2')
ax.set_title('Observation plot')
plt.show()
```

## Logistic regression

```python
# Create a DataFrame of the system
model = pd.DataFrame({'x1' : x1, 'x2' : x2, 'y' : y})
#Selects x1 and x2 as predictors and y as target
features = model.columns.drop('y')
X = MS(features).fit_transform(model)
y = model['y']
```

```python
# Performs a Logistic regression of the entire dataset, samples are drawn from a binomial distribution that corresponds to a logistic loss
glm = sm.GLM(y, X, family=sm.families.Binomial())
log_results = glm.fit()
summarize(log_results)
```

```python
# Split the dataset into a training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Performs a Logistic regression with the train dataset
glm = sm.GLM(y_train, X_train, family=sm.families.Binomial())
log_results = glm.fit()
# Predicted values will be in between 0 and 1 --> probabilities
probs = log_results.predict(X_test)

#Need to convert the continuos predictions into boolean values to confront them with y_test
y_pred_test = np.zeros(len(X_test))
# If prob is larger than 0.5 the result is predicted to be above the median
y_pred_test[probs>0.5] = True
#y_pred_log[probs<0.5] = False

summarize(log_results)
```

```python
y_pred_test
```

```python
probs_train = log_results.predict(X_train)
#Need to convert the continuos predictions into boolean values to confront them with y_test
y_pred_train = np.zeros(len(X_train))
# If prob is larger than 0.5 the result is predicted to be above the median
y_pred_train[probs_train>0.5] = True
confusion_table(y_pred_train, y_train)
```

```python
np.mean(y_pred_train == y_train)
```

```python
confusion_table(y_pred_test, y_test)
```

```python
np.mean(y_pred_test == y_test)
```

```python
fig , ax = plt.subplots(figsize=(8,8))
ax.scatter(X_train['x1'],
X_train['x2'],
c=y_pred_train,
cmap=plt.cm.coolwarm, edgecolor = 'k')
ax.set_xlabel('Train Feature x1')
ax.set_ylabel('Train Feature x2')
ax.set_title('Logistic regression predictions')
plt.show()
```

```python
# Get coefficients
intercept, coef_x1, coef_x2 = log_results.params

# Compute decision boundary
x1_vals = np.linspace(X_train['x1'].min(), X_train['x1'].max(), 100)
x2_vals = -(intercept + coef_x1 * x1_vals) / coef_x2

# Plot with boundary
fig, ax = plt.subplots(figsize=(8, 8))
ax.scatter(X_train['x1'], X_train['x2'], c=y_pred_train, cmap=plt.cm.coolwarm, edgecolors='k')
ax.plot(x1_vals, x2_vals, color='black', linewidth=2, label='Decision Boundary')
ax.set_xlabel('Feature x1')
ax.set_ylabel('Feature x2')
ax.set_title('Training Data with Decision Boundary')
ax.set_ylim([-0.6, 0.6])
ax.legend()
plt.show()
```

## Linear SVM

SVM provides a different approach to classifications wrt logistic regression. --> Different loss

![logvssvm.png](./logvssvm.PNG)
![losses.png](./losses.PNG)

```python
X = np.column_stack((x1,x2))
y = x1**2 - x2**2 > 0
```

```python
# Spilt the dataset into train and test set
(X_train, X_test, y_train, y_test) = skm.train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
svm_lin = SVC(C=10, kernel='linear')
svm_lin.fit(X_train, y_train)
```

```python
kfold = skm.KFold(5, random_state=0, shuffle=True)
grid = skm.GridSearchCV(svm_lin , {'C':[0.1,1,10,100,1000]}, refit=True , cv=kfold , scoring='accuracy');
grid.fit(X_train , y_train)
grid.best_params_
```

```python
# Optimizes the SVC 
svm_lin = SVC(kernel="linear",  C=grid.best_params_['C'])
svm_lin.fit(X_train , y_train)
fig , ax = plt.subplots(figsize=(8,8))
plot_svm(X_train, y_train, svm_lin, scatter_cmap=plt.cm.tab10, ax=ax)
```

```python
best_svm = grid.best_estimator_
#computes test predictions
y_hat_test = best_svm.predict(X_test)
confusion_table(y_hat_test , y_test)
```

```python
np.mean(y_hat_test==y_test)
```

## Poly SVM

Still performing an hinge loss minimization but now the dependence on x is non linear

```python
svm_poly = SVC(C=10, kernel='poly', degree = 2)
svm_poly.fit(X_train, y_train)
```

```python
kfold = skm.KFold(5, random_state=0, shuffle=True)
grid = skm.GridSearchCV(svm_poly , {'C':[0.1,1,10,100,1000], 'gamma':[0.5,1,2,3,4]}, refit=True , cv=kfold , scoring='accuracy');
grid.fit(X_train , y_train)
grid.best_params_
```

```python
# Optimizes the SVC 
svm_poly = SVC(kernel="poly", degree = 2,  C=grid.best_params_['C'], gamma = grid.best_params_['gamma'])
svm_poly.fit(X_train , y_train)
fig , ax = plt.subplots(figsize=(8,8))
plot_svm(X_train, y_train, svm_poly, scatter_cmap=plt.cm.tab10, ax=ax)
```

```python
best_svm = grid.best_estimator_
#computes test predictions
y_hat_test = best_svm.predict(X_test)
confusion_table(y_hat_test , y_test)
```

```python
np.mean(y_hat_test==y_test)
```
