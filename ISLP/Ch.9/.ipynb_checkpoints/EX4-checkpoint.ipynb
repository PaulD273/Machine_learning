```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.model_selection as skm
from sklearn.datasets import make_moons, make_circles
from ISLP import load_data , confusion_table
```

```python
from sklearn.svm import SVC
from ISLP.svm import plot as plot_svm
from sklearn.metrics import RocCurveDisplay
#An ROC curve (Receiver Operating Characteristic curve) is a graphical representation used to evaluate the performance of a binary classification model. 
#It shows the trade-off between the true positive rate (TPR) and the false positive rate (FPR) at various threshold settings.
roc_curve = RocCurveDisplay.from_estimator # shorthand
```

![ex4.png](./ex4.PNG)

```python
X, y = make_circles(n_samples=100, noise=0.1, random_state=42)
#X, y = make_moons(n_samples=100, noise=0.1, random_state=42)
```

```python
#X,  y
```

```python
fig , ax = plt.subplots(figsize=(8,8))
ax.scatter(X[:,0],
X[:,1],
c=y,
cmap=plt.cm.coolwarm)
plt.show()
```

```python
# Spilt the dataset into train and test set
(X_train, X_test, y_train, y_test) = skm.train_test_split(X, y, test_size=0.3, random_state=0)
```

## Degree 3 kernel polynomial

```python
# Perform a Support Vector Classification using a polynomial kernel of order 3 
svm_poly = SVC(kernel="poly", degree = 3, gamma=1, C=1)
svm_poly.fit(X_train , y_train)
```

```python
kfold = skm.KFold(5, random_state=0, shuffle=True)
grid = skm.GridSearchCV(svm_poly , {'C':[0.1,1,10,100,1000], 'gamma':[0.5,1,2,3,4]}, refit=True , cv=kfold , scoring='accuracy');
grid.fit(X_train , y_train)
grid.best_params_
```

```python
# Optimizes the SVC 
svm_poly = SVC(kernel="poly", degree = 3,  C=grid.best_params_['C'], gamma = grid.best_params_['gamma'])
svm_poly.fit(X_train , y_train)
fig , ax = plt.subplots(figsize=(8,8))
plot_svm(X_train, y_train, svm_poly, scatter_cmap=plt.cm.tab10, ax=ax)
```

```python
best_svm = grid.best_estimator_
#computes test predictions
y_hat_test = best_svm.predict(X_test)
confusion_table(y_hat_test , y_test)
```

```python
np.mean(y_hat_test == y_test)
```

```python
#computes test predictions
y_hat_train = best_svm.predict(X_train)
confusion_table(y_hat_train , y_train)
```

```python
np.mean(y_hat_train == y_train)
```

## Radial kernel

```python
# Perform a Support Vector Classification using a radial kernel
svm_rad = SVC(kernel="rbf", C=1, gamma = 1)
svm_rad.fit(X_train , y_train)
```

```python
kfold = skm.KFold(5, random_state=0, shuffle=True)
grid_rad = skm.GridSearchCV(svm_rad , {'C':[0.1,1,10,100,1000], 'gamma':[0.5,1,2,3,4]}, refit=True , cv=kfold , scoring='accuracy');
grid_rad.fit(X_train , y_train)
grid_rad.best_params_
```

```python
# Optimizes the SVC 
svm_rad = SVC(kernel="rbf", C=grid_rad.best_params_['C'], gamma = grid_rad.best_params_['gamma'])
svm_rad.fit(X_train , y_train)
fig , ax = plt.subplots(figsize=(8,8))
plot_svm(X_train, y_train, svm_rad, scatter_cmap=plt.cm.tab10, ax=ax)
```

```python
best_svm_rad = grid_rad.best_estimator_
#computes test predictions
y_hat_test_rad = best_svm_rad.predict(X_test)
confusion_table(y_hat_test_rad , y_test)
```

```python
np.mean(y_hat_test_rad == y_test)
```

```python
#computes test predictions
y_hat_train_rad = best_svm_rad.predict(X_train)
confusion_table(y_hat_train_rad , y_train)
```

```python
np.mean(y_hat_train_rad == y_train)
```

With a radially separable two class dataset the radial kernel performs better with a roughly 83% accuracy in the test set, followed by the degree 3 polynomial with a roughly 60% accuracy and trailing is the linear kernel that with a 50% accuracy is as good as random guessing.

With a Moon dataset the scores are much closer, due to the parabolic distribution of the features the linear models has a decent performance, but the radial kernel still proves to be the most apt.
