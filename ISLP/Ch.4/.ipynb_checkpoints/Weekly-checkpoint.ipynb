## Part b

```python
# The 5 lag variables plus volume are used as predictors
allvars = weekly.columns.drop(['Direction', 'Year', 'Today'])
X = MS(allvars).fit_transform(weekly)
# Converts direction into a boolean variable
y = weekly.Direction == 'Up'
# Performs a Logistic regression
glm = sm.GLM(y, X, family=sm.families.Binomial())
log_results = glm.fit()
summarize(log_results)
```

```python
# Prediction probability of the direction
log_probs = log_results.predict()
#Since direction can only have a binary set of values, if the prob is larger than 0.5 is set to up otherwise it is set to down
log_labels = np.array(['Down']*len(weekly))
log_labels[log_probs >0.5] = "Up"
confusion_table(log_labels, weekly.Direction)
```

```python
np.mean(log_labels == weekly.Direction)
```

56% of success on the training set, no out of set predictions are included. Almost equal probability of correctly predicting an Up day.

All predictors are, strictly applying the 0.05 rule, statistically insignificant, though Lag1 and Lag2 are less insignificant than the others

```python
# Only Lag1 and Lag2 are used as predictors
Lag12 = weekly[['Lag1', 'Lag2']]
X12 = MS(Lag12).fit_transform(weekly)
# Converts direction into a boolean variable
y = weekly.Direction == 'Up'
# Performs a Logistic regression
glm = sm.GLM(y, X12, family=sm.families.Binomial())
log_results12 = glm.fit()
summarize(log_results12)
```

```python
# Prediction probability of the direction
log_probs12 = log_results12.predict()
#Since direction can only have a binary set of values, if the prob is larger than 0.5 is set to up otherwise it is set to down
log_labels12 = np.array(['Down']*len(weekly))
log_labels12[log_probs12 >0.5] = "Up"
confusion_table(log_labels12, weekly.Direction)
```

```python
np.mean(log_labels == weekly.Direction)
```

Identical probability of success.

```python
# The market is assumed to be rising everyday
labels_naive = np.array(['Up']*len(weekly))
confusion_table(labels_naive , weekly.Direction)
```

```python
np.mean(labels_naive == weekly.Direction)
```

By assuming the market to go Up each day we basically have the same probability of success of the model.

## Part d

```python
# Observations up to the year 2008 are set to True, the others are set to false
train = (weekly.Year < 2009)
# Trains using pre 2009 data of the Lag2 predictor
weekly_train = weekly[['Lag2']].loc[train]
# Tests with 2009 & 2010 data of the Lag2 predictor
weekly_test = weekly[['Lag2']].loc[~train]
# Includes only the Lag2 parameter among the predictors
X = MS(weekly[['Lag2']]).fit_transform(weekly)
# Splits the X dataframe (which excludes year, today and direction) into test and training
X_train , X_test = X.loc[train], X.loc[~train]
y_train , y_test = y.loc[train], y.loc[~train]
```

```python
weekly_train, weekly_test
```

```python
# Perform a logistic regression on the training set
glm_train = sm.GLM(y_train , X_train , family=sm.families.Binomial())
results = glm_train.fit()
# Predicts using the test set
probs = results.predict(exog=X_test)
```

```python
probs
```

```python
results.summary()
```

```python
# Extracts the direction column for the training and test
L_train , L_test = weekly.Direction.loc[train], weekly.Direction.loc[~train]
# Converts the predictions into Up and Down strings
log_pred = np.array(['Down']*len(X_test))
log_pred[probs >0.5] = 'Up'
# Computes the confusion table
confusion_table(log_pred , L_test)
```

```python
log_rate = np.mean(log_pred == L_test)
```

62.5% success rate on the test set, higher than that obtained using all 5 lags and volume as predictors.

## Part e

```python
# Perform an LDA classification on the training set, using the same training and test set as before
lda = LDA(store_covariance=True)
lda.fit(X_train , L_train)
```

```python
lda_pred = lda.predict(X_test)
confusion_table(lda_pred , L_test)
```

```python
lda_rate = np.mean(lda_pred == L_test)
```

62.5% success rate on the test set, same performance as the logistic regression

## Part f

```python
# Perform a QDA classification on the training set, using the same training and test set as before
qda = QDA(store_covariance=True, reg_param=0.1)
qda.fit(X_train , L_train)
```

```python
qda_pred = qda.predict(X_test)
confusion_table(qda_pred, L_test)
```

```python
qda_rate = np.mean(qda_pred == L_test)
```

## Part g

```python
# Perform a KNN1 classification on the training set, using the same training and test set as before
knn1 = KNeighborsClassifier(n_neighbors=1)
knn1.fit(X_train , L_train)
```

```python
knn1_pred = knn1.predict(X_test)
confusion_table(knn1_pred , L_test)
```

```python
knn1_rate = np.mean(knn1_pred == L_test)
```

## Part h

```python
# Perform a Naive Bayes classification on the training set, using the same training and test set as before
NB = GaussianNB()
NB.fit(X_train , L_train)
```

```python
NB_pred = NB.predict(X_test)
confusion_table(NB_pred , L_test)
```

```python
NB_rate = np.mean(NB_pred == L_test)
```

## Part i

```python
success_rates = {'Logistic' : round(log_rate,3),
                             'LDA' : round(lda_rate,3),
                             'QDA' : round(qda_rate,3),
                             'KNN1' : round(knn1_rate,3),
                             'Naive Bayes' : round(NB_rate,3)}
```

```python
for method, score in success_rates.items():
    print(f"{method}: Score = {score}")
```
