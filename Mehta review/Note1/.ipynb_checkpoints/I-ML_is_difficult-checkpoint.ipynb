# Notebook 1: Why is Machine Learning difficult?

## Overview 

In this notebook, we will get our hands dirty trying to gain intuition about why machine learning is difficult. 

Our task is going to be a simple one, fitting data with polynomials of different order. Formally, this goes under the name of polynomial regression. Here we will do a series of exercises that are intended to give the reader intuition about the major challenges that any machine learning algorithm faces.

## Learning Goal

We will explore how our ability to predict depends on the number of data points we have, the "noise" in the data, and our knowledge about relevant features. The goal is to build intuition about why prediction is difficult and discuss general strategies for overcoming these difficulties.


## The Prediction Problem

Consider a probabilistic process that gives rise to labeled data $(x,y)$. The data is generated by drawing samples from the equation
$$
    y_i= f(x_i) + \eta_i,
$$
where $f(x_i)$ is some fixed, but (possibly unknown) function, and $\eta_i$ is a Gaussian, uncorrelate noise variable such that
$$
\langle \eta_i \rangle=0 \\
\langle \eta_i \eta_j \rangle = \delta_{ij} \sigma
$$
We will refer to the $f(x_i)$ as the **true features** used to generate the data. 

To make prediction, we will consider a family of functions $g_\alpha(x;\theta_\alpha)$ that depend on some parameters $\theta_\alpha$. These functions respresent the **model class** that we are using to try to model the data and make predictions. The $g_\alpha(x;\theta_\alpha)$ encode the class of **features** we are using to represent the data.

To learn the parameters $\boldsymbol{\theta}$, we will train our models on a **training data set** and then test the effectiveness of the model on a <i>different</i> dataset, the **test data set**. The reason we must divide our data into a training and test dataset is that the point of machine learning is to make accurate predictions about new data we have not seen. As we will see below, **models that give the best fit to the training data do not necessarily make the best predictions on the test data**. This will be a running theme that we will encounter repeatedly in machine learning.  


For the remainder of the notebook, we will focus on polynomial regression. Our task is to model the data with polynomials and make predictions about the new data that we have not seen.
We will consider two qualitatively distinct situations: 
<ul>
<li> In the first case, the process that generates the underlying data is in the model class we are using to make predictions. For polynomial regression, this means that the functions $f(x_i)$ are themselves polynomials.
<li>In the second case, our data lies outside our model class. In the case of polynomial regression, this could correspond to the case where the $f(x_i)$ is a 10-th order polynomial but $g_\alpha(x;\theta_\alpha)$ are polynomials of order 1 or 3.
</ul>

In the exercises and discussion we consider 3 model classes:
<ul>
<li> the case where the $g_\alpha(x;\theta_\alpha)$ are all polynomials up to order 1 (linear models),
<li> the case where the $g_\alpha(x;\theta_\alpha)$ are all polynomials up to order 3,
<li> the case where the $g_\alpha(x;\theta_\alpha)$ are all polynomials up to order 10.
</ul>

To measure our ability to predict, we will learn our parameters by fitting our training dataset and then making predictions on our test data set. One common measure of predictive  performance of our algorithm is to compare the predictions,$\{y_j^\mathrm{pred}\}$, to the true values $\{y_j\}$. A commonly employed measure for this is the sum of the mean square-error (MSE) on the test set:
$$
MSE= \frac{1}{N_\mathrm{test}}\sum_{j=1}^{N_\mathrm{test}} (y_j^\mathrm{pred}-y_j)^2
$$
We will return to this in later notebooks. For now, we will try to get a qualitative picture by examining plots on test and training data.

# Training the models:

```python
#This is Python Notebook to walk through polynomial regression examples
#We will use this to think about regression
import numpy as np
%matplotlib inline

from sklearn import datasets, linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

from matplotlib import pyplot as plt, rcParams
fig = plt.figure(figsize=(8, 6))

# The Training Data
N_train=10
sigma_train=1

#np.random.seed(49)
# Train on integers
x=np.linspace(0, 1,N_train)
# Draw Gaussian random noise
s = sigma_train*np.random.randn(N_train)

#linear
#y=2*x+s

# Tenth Order
y=2*x-10*x**5+15*x**10+s

p1=plt.plot(x, y, "o", ms=8, alpha=0.5, label='Training')

# Linear Regression : create linear regression object
clf = linear_model.LinearRegression()

# Train the model using the training set
# Note: sklearn requires a design matrix of shape (N_train, N_features). Thus we reshape x to (N_train, 1):
clf.fit(x[:, np.newaxis], y)
# np.newaxis is a convenient alias for None, useful for indexing arrays.

# Use fitted linear model to predict the y value:
xplot=np.linspace(0.02,0.98,200) # grid of points, some are in the training set, some are not
linear_plot=plt.plot(xplot, clf.predict(xplot[:, np.newaxis]), label='Linear')

# Polynomial Regression
# Creates a transformer that will generate polynomial features up to degree 3
poly3 = PolynomialFeatures(degree=3)
# Construct polynomial features
# Then, fit_transform constructs the polynomial feature matrix. For each element x_i, youll now have
# a row like [1, x_i, x_i**2, x_i**3]
X = poly3.fit_transform(x[:,np.newaxis])
#Instantiates a linear regression model.
clf3 = linear_model.LinearRegression()
# Fits the model using the transformed polynomial features X and the target variable y.
clf3.fit(X,y)


Xplot=poly3.fit_transform(xplot[:,np.newaxis])
poly3_plot=plt.plot(xplot, clf3.predict(Xplot), label='Poly 3', color = 'blue', linestyle='-.')

# Fifth order polynomial in case you want to try it out
#poly5 = PolynomialFeatures(degree=5)
#X = poly5.fit_transform(x[:,np.newaxis])
#clf5 = linear_model.LinearRegression()
#clf5.fit(X,y)

#Xplot=poly5.fit_transform(xplot[:,np.newaxis])
#plt.plot(xplot, clf5.predict(Xplot), 'r--',linewidth=1)

poly10 = PolynomialFeatures(degree=10)
X = poly10.fit_transform(x[:,np.newaxis])
clf10 = linear_model.LinearRegression()
clf10.fit(X,y)

Xplot=poly10.fit_transform(xplot[:,np.newaxis])
poly10_plot=plt.plot(xplot, clf10.predict(Xplot), label='Poly 10', linestyle='--')

# Linear
#plt.plot(xplot, 2*xplot, label = 'model')
# Tenth order
plt.plot(xplot, 2*xplot-10*xplot**5+15*xplot**10, label = 'model')

plt.legend(loc='best')
plt.ylim([-7,7])
plt.xlabel("x")
plt.ylabel("y")
Title="N=%i, $\sigma=%.2f$"%(N_train,sigma_train)
plt.title(Title+" (train)")


# Linear Filename
#filename_train="train-linear_N=%i_noise=%.2f.png"%(N_train, sigma_train)

# Tenth Order Filename
filename_train="train-o10_N=%i_noise=%.2f.png"%(N_train, sigma_train)

# Saving figure and showing results
#plt.savefig(filename_train)
plt.grid()
plt.show()
```

```python
# Slope predicted by the linear model
clf.fit(x[:, np.newaxis], y).coef_[0]
```

```python
# Recompute in-sample error using actual training data
X_train_poly1 = x[:, np.newaxis]
X_train_poly3 = poly3.transform(X_train_poly1)
X_train_poly10 = poly10.transform(X_train_poly1)

train_error_lin = mean_squared_error(y, clf.predict(X_train_poly1))
train_error_3 = mean_squared_error(y, clf3.predict(X_train_poly3))
train_error_10 = mean_squared_error(y, clf10.predict(X_train_poly10))

print("In-sample Errors:")
print("Linear:", train_error_lin)
print("Degree 3:", train_error_3)
print("Degree 10:", train_error_10)
```

# Testing the fitted models

```python
# Generate Test Data
%matplotlib inline
# Number of test data
N_test=20
sigma_test=sigma_train

# Generate random grid points (x) in the interval [0, max_x]:
# Note some points will be drawn outside the training interval
max_x=1.2
x_test=max_x*np.random.random(N_test)

# Draw random Gaussian noise
s_test = sigma_test*np.random.randn(N_test)

# Linear
#y_test=2*x_test+s_test
# Tenth order
y_test=2*x_test-10*x_test**5+15*x_test**10+s_test

# Make design matrices for prediction
x_plot=np.linspace(0,max_x, 200)
X3 = poly3.fit_transform(x_plot[:,np.newaxis])
X10 = poly10.fit_transform(x_plot[:,np.newaxis])

############## PLOTTING RESULTS ##########

fig = plt.figure(figsize=(8, 6))

p1=plt.plot(x_test, y_test, 'o', ms=10, alpha=0.5, label='test data')
p2=plt.plot(x_plot,clf.predict(x_plot[:,np.newaxis]), lw=2, label='linear')
p3=plt.plot(x_plot,clf3.predict(X3), lw=2, label='3rd order', color = 'blue', linestyle='-.')
p10=plt.plot(x_plot,clf10.predict(X10), lw=2, label='10th order', linestyle='--')

# Linear
#plt.plot(x_plot, 2*x_plot, label = 'model')
# Tenth order
plt.plot(x_plot, 2*x_plot-10*x_plot**5+15*x_plot**10, label = 'model')

plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='best')
Title="N=%i, $\sigma=%.2f$"%(N_test,sigma_test)
plt.title(Title+" (pred.)")
plt.tight_layout()
plt.ylim((-6,12))

# Linear Filename
#filename_test="pred-linear_N=%i_noise=%.2f.png"%(N_test, sigma_test)

# Tenth Order Filename
filename_test="pred-o10_N=%i_noise=%.2f.png"%(N_test, sigma_test)

# Saving figure and showing results
#plt.savefig(filename_test)
plt.grid()
plt.show()
```

```python
X_test_poly1 = x_test[:, np.newaxis]
X_test_poly3 = poly3.transform(X_test_poly1)
X_test_poly10 = poly10.transform(X_test_poly1)

test_error_lin = mean_squared_error(y_test, clf.predict(X_test_poly1))
test_error_3 = mean_squared_error(y_test, clf3.predict(X_test_poly3))
test_error_10 = mean_squared_error(y_test, clf10.predict(X_test_poly10))

print("\nOut-of-sample Errors:")
print("Linear:", test_error_lin)
print("Degree 3:", test_error_3)
print("Degree 10:", test_error_10)
```

## Case $N_{train}$ = 10 and $\sigma = 0$

![train-linear_N=10_noise=0.00.png](./train-linear_N=10_noise=0.00.png)
![pred-linear_N=20_noise=0.00.png](./pred-linear_N=20_noise=0.00.png)

In the absence of any noise ($\sigma = 0$) and with a limited training set ($N_{train} = 10$), all three models fit the function perfectly on both the training and test set. The prediction is perfect for features both inside the training interval ($ x \in \left[ 0, 1 \right]$) and outside it (e.g. $ x \in \left[ 0, 1.2 \right]$).

Simpler models are usually preferred when the function is simple, high degree polynomials may lead to oscillations outside the training range, even though the function is fitted perfectly within it (Runge oscillations).

![Runge_oscillation.jpg](./Runge_oscillation.jpg)

Good training fit does not guarantee good prediction. Simple models, even if imperfect on training data, can yield better predictions because they capture the true underlying nature behind the noise. Complex models may fit exactly training data but the have terrible test accuracy.

This mechanism can be explained by the so called **Bias-Variance tradeoff**

![bias_var.jpg](./bias_var.jpg)

## Case $N_{train}$ = 10 and $\sigma = 1.00$

![train-linear_N=10_noise=1.00.png](./train-linear_N=10_noise=1.00.png)
![pred-linear_N=20_noise=1.00.png](./pred-linear_N=20_noise=1.00.png)

**The introduction of noise ($\sigma = 1$) exposes overfitting in complex models**. The linear model still fits the overall trend (slope = 2.3309309968599425 $\sim$ 2) and it does not chase individual noisy points. The third order polynomial slightly overfits the dataset while the tenth order polynomial massively overfits it, trying to match some noise fluctuations.

Outside the training set (e.g. $ x \in \left[ 0, 1.2 \right]$), the overfitting is amplified. 

Simple models generalize better when the data is noisy.

With a much larger dataset ($N_{train} = 100$) overfitting is reduced, noise averages out with more data points. All models predict test data better with more training points, even though some oscillations are still visibile outside the training range for the tenth order function. 

**Increasing the amount of data reduces overfitting**. 

**Complex models need more data** to fit safely without overfitting.

Without noise and with few data points, the linear model is uncapable to mimic the complexity of the real function (order 10), the same goes for the third order polynomial (they are both affected by high bias). The tenth order polynomial, on the other hand, matches the true function exactly both inside and outside the training interval, it is thus affected by a very low bias.

As expected, if the true function lies within the model class, then with noiseless data, the model can recover it exactly.

## Case $N_{train}$ = 10 and $\sigma = 1.00$

![train-o10_N=10_noise=1.00.png](./train-o10_N=10_noise=1.00.png)
![pred-o10_N=30_noise=1.00.png](./pred-o10_N=30_noise=1.00.png)

Adding some noise the fitting unravels. The linear model is not much affected by noise but still underfits the real function, the third order polynomial starts to overfit noise slightly and it still cannot capture the full shape of the function, the tenth order polynomial overfits badly the noise and test performance is poor, especially outside the training region, where we have wild oscillations (affected by high variance, very low training error but terrible generalization). 

**Models that give the best fit to the training data do not necessarily make the best predictions on the test data**.

With more data, the noise gets averaged out and the tenth order function is able to fit the true function without overfitting noise too much. Prediction is also improved but we still have oscillations outside the training range.

**Complex models need large datasets to generalize well**, otherwise they just memorize noise.
